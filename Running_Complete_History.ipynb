```python
# libraries for data analysis and graphing
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#https://docs.google.com/spreadsheets/d/1YqCft96bmlRZd1k59PzchDOIbY3vJE7tlRFb66rwVn0/edit?usp=sharing
df_sheet_id = '1YqCft96bmlRZd1k59PzchDOIbY3vJE7tlRFb66rwVn0'
df_sheet_name = 'running_complete'
df_url = "https://docs.google.com/spreadsheets/d/{}/gviz/tq?tqx=out:csv&sheet={}".format(df_sheet_id,df_sheet_name)
df = pd.read_csv(df_url, parse_dates = ['Datetime'])
df.head()
df['month_year'] = df.Datetime.dt.to_period('M')
df['quarter'] = df.Datetime.dt.to_period('Q')
df['year'] = df.Datetime.dt.to_period('Y')
```

```python
plt.figure(figsize=(15,10))
chart = sns.boxplot(data=df,
    x='quarter', y='Avg_Speed', color='#9BE5B4').set(title='Complete Running History')
#chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
plt.xticks(
    rotation=90,
    horizontalalignment='right',
    fontweight='light',
    fontsize='small'
)
plt.xlabel("quarter")
plt.ylabel("miles per hour")
plt.savefig('quarterly_running_speed_history.png')
```

```python
# color can be set to hex code

plt.figure(figsize=(15,10))
chart = sns.boxplot(data=df,
    x='quarter', y='Miles', color='#9BE5B4').set(title='Complete Running History')
#chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
plt.xticks(
    rotation=90,
    horizontalalignment='right',
    fontweight='light',
    fontsize='small'
)
plt.xlabel("quarter")
plt.ylabel("running distance (miles)")
plt.savefig('quarterly_running_distance_history.png')
```
