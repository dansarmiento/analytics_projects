```python
# Send a GET request to the API endpoint
response = requests.get(api_url)
```

```python
# medicare physician database https://data.cms.gov/data-api/v1/dataset/92396110-2aed-4d63-a6a2-5d6207d46a29/data
# Base URL for the API endpoint
base_url = "https://data.cms.gov/data-api/v1/dataset/92396110-2aed-4d63-a6a2-5d6207d46a29/data"
all_data = []

# Parameters for pagination
params = {
    'page': 1,
    'page_size': 1000  # Adjust the page size as necessary
}

all_data = []

while True:
    response = requests.get(base_url, params=params)
    if response.status_code == 200:
        # Parse the JSON response
        data = response.json()
        if not data:
            break
        all_data.extend(data)

        # Check if more data is available
        params['page'] += 1
    else:
        print(f"Failed to retrieve data. HTTP Status code: {response.status_code}")
        break

```
